```python
#### Data cleaning and preprocessing 

import nltk 
import re   ## regular expresion
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
lemma
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', str(messages.iloc[i, 1]))  # Convert to string if it's NaN
    review = review.lower()
    review = review.split()
    review =  [lemma.lemmatize(word)for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#####  for binary bag of word enable binary = Ture 

cv = CountVectorizer(max_features=15, binary = True)
```

```python
y = cv.fit_transform(corpus).toarray()
```
